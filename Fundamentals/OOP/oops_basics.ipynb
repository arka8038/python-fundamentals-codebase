## Instance Attributes

````python
## Creating a class - instantiating objects with attributes

## 'self' keyword is a reference to the current instance of the class
## 'self' is used to access variables that belong to the class
class User:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
user1 = User("John", 30)
user2 = User("Jane", 25)
user3 = User("Doe", 40)

print(user1.name, user1.age)  # John 30
print(user2.name, user2.age)  # Jane 25
print(user3.name, user3.age)  # Doe 40    
````

## Dunder Naming Conventions in Python

In Python, names with underscores have special meanings and conventions. Here's a breakdown of the three main types:

---

### `_name` (Single Leading Underscore)

- **Convention only**: Not enforced by the language.
- Indicates the attribute or method is intended for internal use only.
- When using `from module import *`, names with a leading underscore are not imported.
- **Example:**
  ```python
  class MyClass:
      def __init__(self):
          self._internal_var = 42
  ```

---

### `__name` (Double Leading Underscore)

- Triggers **name mangling**: Python internally renames the attribute to `_ClassName__name`.
- Used to avoid accidental access or override in subclasses.
- Not for strict access control but for preventing name collisions.
- **Example:**
  ```python
  class MyClass:
      def __init__(self):
          self.__private_var = 42
  ```

---

### `__name__` (Double Leading and Trailing Underscores)

- Reserved for **special methods or variables** defined by Python.
- Commonly called **"dunder methods"** (double underscore).
- Used to implement behavior or configuration, like `__init__`, `__str__`, or `__name__`.
- You should **not create your own** names like thisâ€”only use existing Python-defined dunder names.
- **Example:**
  ```python
  if __name__ == "__main__":
      print("Running as a script")
  ```

## Instance Methods in Python

Instance methods are functions defined inside a class that operate on instances of that class.

---

### ðŸ”¹ Key Features

- Always take `self` as the **first parameter**.
- Can access and modify **instance attributes** and call other instance methods.
- Called on an **instance** of the class.

---

### âœ… Example

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

    def start_engine(self):
        print(f"{self.brand} engine started.")

my_car = Car("Toyota")
my_car.start_engine()  # Output: Toyota engine started.


````python
class User:
    def __init__(self, first, last, age):
        self.first = first
        self.last = last
        self.age = age
        
    def likes(self, things):
        return f"{self.first} likes {things}" 
    
    
    def is_senior_citizen(self):
        return self.age >= 65
    
    ## method to increment age and return a birthday message
    ## 'self' is used to access the instance variable 'age' and increment it by 1
    def birthday(self):
        self.age += 1
        return f"Happy Birthday {self.first} {self.last}, it is your {self.age}th birthday!"
    
u1 = User("Joe", "Smith", 60)       
print(u1.likes('python')) # Joe likes python
print(u1.is_senior_citizen())  # False
print(u1.birthday())  # Happy Birthday Joe Smith, it is your 61th birthday!
print(u1.age)
````

## Class Attributes

Class attributes are variables that are shared across all instances of a class. They are defined directly inside the class body, outside any instance methods.

---

### ðŸ”¹ What Are Class Attributes?

- Belong to the **class itself**, not any one instance.
- Shared by **all instances** of the class.
- Typically used for constants or values common to all objects of the class.

---

### ðŸ§  When to Use Class Attributes

- When you need to share a value across all instances.
- To define class-level constants (e.g., `MAX_SIZE = 100`).
- To track or manage class-level state (like instance counters).

---

### âœ… Example

```python
class Dog:
    # Class attribute
    species = "Canis familiaris"

    def __init__(self, name):
        self.name = name  # Instance attribute

dog1 = Dog("Buddy")
dog2 = Dog("Lucy")

print(dog1.species)  # Output: Canis familiaris
print(dog2.species)  # Output: Canis familiaris
```

---

### âš ï¸ Important Notes

- Class attributes can be accessed via both the class and the instance.
- If you **assign** a value to the attribute on an instance, it **creates an instance attribute**, shadowing the class attribute.

```python
dog1.species = "Different species"
print(dog1.species)  # Output: Different species
print(dog2.species)  # Output: Canis familiaris
```

---

### ðŸ†š Class Attribute vs Instance Attribute

| Feature              | Class Attribute            | Instance Attribute         |
|----------------------|-----------------------------|-----------------------------|
| Defined in           | Class body                  | Inside `__init__()` or methods |
| Shared across        | All instances               | Unique to each instance     |
| Changed by           | Class or instance (shadows) | Each individual instance     |

````python
class User:
    active_user = 0  # class variable to keep track of active users
    
    def __init__(self, first, last, age):
        self.first = first 
        self.last = last
        self.age = age  ## these are for instance variable
        
        User.active_user += 1  # increment active user count when a new user is created
        
    def logout(self):
        User.active_user -= 1
        return f"{self.first} {self.last} has logged out."
    
# print(User.__dict__)  # shows all the attributes of the class including class variables 

print(User.active_user) # 0 active users

### creating instances of User class
u1 = User("Joe", "Smith", 60) 
u2 = User("Jane", "Doe", 25)
u3 = User("John", "Doe", 40)

print(User.active_user) #  # 3 active users

print(u1.logout())  # Joe Smith has logged out.
print(u2.logout())  # Jane Doe has logged out.

print(User.active_user) # 1 active users
````

````python
class Pet:
    allowed = ["cat", "dog", "fish"]  # class variable to keep track of allowed pet types
    
    def __init__(self, name, species):
        if species not in Pet.allowed: ## can also use self.allowed but it is better to use class variable
            raise ValueError(f"{species} is not allowed")
        
        self.name = name
        self.species = species
        
p1 = Pet("Fluffy", "cat")
p2 = Pet("Fido", "dog")
p3 = Pet("Nemo", "fish")
# p4 = Pet("Guppy", "goldfish")  ## raises ValueError: goldfish is not allowed      
        
````

## Class Methods

A **class method** is a method that's bound to the class and not the instance. It can access and modify class state that applies across all instances.

---

### ðŸ”¹ Key Characteristics

- Defined using the `@classmethod` decorator.
- Takes `cls` as the first parameter (not `self`).
- Can **access or modify class attributes**, but **cannot access instance attributes directly**.
- Can be called using either the class or an instance.

---

### âœ… Example

```python
class Book:
    book_count = 0  # Class attribute

    def __init__(self, title):
        self.title = title
        Book.increment_count()

    @classmethod
    def increment_count(cls):
        cls.book_count += 1

    @classmethod
    def from_string(cls, book_str):
        # Factory method to create Book from string
        title = book_str.strip()
        return cls(title)

# Using class method as a factory
book1 = Book.from_string("  Python 101  ")
print(book1.title)         # Output: Python 101
print(Book.book_count)     # Output: 1
```

---

### ðŸ§  Why and When to Use Class Methods

| Use Case                         | Why Itâ€™s Useful                                                                 |
|----------------------------------|----------------------------------------------------------------------------------|
| Modify class-level state         | Use when you need to access or change attributes shared by all instances.        |
| Alternative constructors         | Use `@classmethod` to define **factory methods** that return instances.         |
| Maintain consistency             | Class methods ensure changes affect the class itself, not just individual objects. |
| Track number of instances        | Increment a class counter every time an instance is created.                    |

---

### ðŸ†š Class Method vs Instance Method vs Static Method

| Feature             | Instance Method       | Class Method           | Static Method             |
|---------------------|------------------------|--------------------------|----------------------------|
| First arg           | `self`                | `cls`                   | None                       |
| Access instance data| âœ…                    | âŒ                      | âŒ                         |
| Access class data   | âœ… (via instance)     | âœ…                      | âŒ                         |
| Use case            | Object behavior       | Factory methods, shared data | Utility/helper functions   |

---

### ðŸ”§ Real-World Example: Factory Method

```python
class User:
    def __init__(self, username, email):
        self.username = username
        self.email = email

    @classmethod
    def from_dict(cls, data):
        return cls(data["username"], data["email"])

data = {"username": "john_doe", "email": "john@example.com"}
user = User.from_dict(data)
```

---

### ðŸ“Œ Summary

- Use `@classmethod` when the method needs to know or affect the class itself.
- Great for factory patterns and managing class-level data.
- Helps in maintaining consistent class-wide behavior across instances.

```

````python
class User:
    active_users = 0  # class variable to keep track of active users
    
    @classmethod
    def display_active_users(cls):
        return f"There are {cls.active_users} active users."
        
    def __init__(self, first, last, age):
        self.first = first
        self.last = last
        self.age = age
        User.active_users += 1  # increment active user count when a new user is created
        

u1 = User("Joe", "Smith", 60)
u2 = User("Jane", "Doe", 25)

print(User.display_active_users())
````

## `__repr__` in Python

The `__repr__` method is a **special method** used to define the **official string representation** of an object.

---

### ðŸ”¹ Purpose

- Returns a string that would ideally be a **valid Python expression** to recreate the object.
- Called by the built-in `repr()` function and used in **debugging**, **logging**, and **interactive sessions** (e.g., in a REPL or Jupyter Notebook).

---

### âœ… Example

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name='{self.name}', age={self.age})"

p = Person("Alice", 30)
print(repr(p))  # Output: Person(name='Alice', age=30)
```

---

### ðŸ§  Why Use `__repr__`?

| Situation                        | Reason                                                   |
|----------------------------------|-----------------------------------------------------------|
| Debugging and logging            | Helps identify object state clearly.                     |
| Interactive Python environments | Controls how objects appear in the console or shell.     |
| Re-creatable objects             | Aim to make the output something that could recreate the object if passed to `eval()` (though this is not always required).

---

### ðŸ”„ `__repr__` vs `__str__`

| Feature     | `__repr__`                                 | `__str__`                            |
|-------------|---------------------------------------------|--------------------------------------|
| Purpose     | Developer/debug-friendly representation     | User-friendly (readable) output      |
| Called by   | `repr(obj)` or in shell                     | `str(obj)` or `print(obj)`           |
| Fallback    | If `__str__` is missing, Python uses `__repr__` | If `__repr__` is missing, default repr is used |

---

### ðŸ§ª Tip

Always implement `__repr__` for **custom classes** â€” it makes debugging and development much easier.
